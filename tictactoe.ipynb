```python
#tic tac toe  k-ply (4-depth)
#there should be a state
from copy import deepcopy
class State:
    #initializing
    def __init__(self,board=None,player='âœ–ï¸'):
        if board is None:
            self.board=[[' ' for _ in range(3)] for _ in range(3)]
            #above is done int the beginning
        else:
            self.board=board
        self.player=player#turn
    def goalTest(self):
        #if any row or diag or col is filled with all X or O
        for i in range(3):
            if self.board[i][0]==self.board[i][1]==self.board[i][2]!=' ':#this row is filled 
                return self.board[i][0]
            if self.board[0][i]==self.board[1][i]==self.board[2][i]!=' ':#this col is filled 
                return self.board[0][i]
            #we need to check for diagnols
        if self.board[1][1]==self.board[2][2]==self.board[0][0]!=' ':
            return self.board[0][0]
        if self.board[0][2]==self.board[1][1]==self.board[2][0]!=' ':
            return self.board[1][1]
        #goal not found
        return None 
    def isFull(self):
        return all(self.board[i][j]!=' ' for i in range(3) for j in range(3))
    
    def moveGen(self):
        children=[]
        for i in range(3):
            for j in range(3):
                if(self.board[i][j]==' '):
                    #we need to fill this
                    newboard=deepcopy(self.board)
                    newboard[i][j]=self.player
                    nextplayer='âœ–ï¸' if self.player=='â¤ï¸â€ğŸ”¥ï¸' else 'â¤ï¸â€ğŸ”¥ï¸'
                    children.append(State(newboard,nextplayer))
                    #simple in empty place they filled 
                    #next states are generated 
        return children
    def __str__(self):#just for ui
        rows=[]
        for row in self.board:
            rows.append("|".join(row))
        return "\n--------\n".join(rows)+"\n"
    
    def __eq__(self, value):
        return self.board==value.board and self.player==value.player
    def __hash__(self):
        return hash(str(self.board))
    def evaluate(self):#points
        #we need to evalute the points 
        #find the best move where ai wins
        winner=self.goalTest()
        if winner=='âœ–ï¸':
            return 10
        elif winner=='â¤ï¸â€ğŸ”¥ï¸':
            return -10
        ai_count,human_count=0,0

        for i in range(3):
            row=self.board[i][:]#full row
            col=[self.board[j][i] for j in range(3)]#full columns

            if 'â¤ï¸â€ğŸ”¥ï¸' not in row:ai_count+=1
            if 'âœ–ï¸' not in row:human_count+=1
            if 'â¤ï¸â€ğŸ”¥ï¸' not in col:ai_count+=1
            if 'âœ–ï¸' not in col:human_count+=1

            d1=[self.board[k][k]for k in range(3)]
            d2=[self.board[k][2-k] for k in range(3)]

            if 'â¤ï¸â€ğŸ”¥ï¸' not in d1:ai_count+=1
            if 'âœ–ï¸' not in d1:human_count+=1
            if 'â¤ï¸â€ğŸ”¥ï¸' not in d2:ai_count+=1
            if 'âœ–ï¸' not in d2:human_count+=1
        return ai_count-human_count

```

```python
def alpha_beta(state,alpha,beta,depth,maxdepth,maxplayer):
    winner=state.goalTest()
    if winner or state.isFull() or depth==maxdepth:
        return state.evaluate(),None #count is returned 
    best_move=None
    if maxplayer:#max node
        v=float('-inf')
        for child in state.moveGen():
            eval_score,_=alpha_beta(child,alpha,beta,depth+1,maxdepth,False)
            if eval_score>v:
                v=eval_score
                best_move=child
            alpha=max(alpha,eval_score)
            #pruning
            if alpha>=beta:break
        return v,best_move
    else:#min node
        v=float('inf')
        for child in state.moveGen():
            eval_score,_=alpha_beta(child,alpha,beta,depth+1,maxdepth,True)
            if eval_score<v:
                v=eval_score
                best_move=child
            beta=min(beta,eval_score)
            #pruning
            if alpha>=beta:break
        return v,best_move
```
